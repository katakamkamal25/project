```python
import cv2

import time
car_classifier = cv2.CascadeClassifier('cars.xml')

cap = cv2.VideoCapture('video3.avi')
fgbg1 = cv2.bgsegm.createBackgroundSubtractorMOG()



ax1= 70
ay=  90
ax2=200
#line b
bx1= 15
by= 125
bx2=225
def Speed_Cal(time):
    try:
        Speed = (9.144*3600)/(time*1000)
        return Speed
    except ZeroDivisionError:
        print (5)
i = 1
start_time = time.time()


while True:
    ret, frame = cap.read()
    if (type(frame) == type(None)):
        break

    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
   
   

    cars = car_classifier.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3, minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)
    
   
    

    for (x, y, w, h) in cars:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)               

        cv2.circle(frame,(int((x+x+w)/2),int((y+y+h)/2)),5,(0,255,0),-1)   
        while int(ay) == int((y+y+h)/2):
            start_time = time.time()
            break
        cv2.line(frame,(ax1,ay),(ax2,ay),(255,0,0),2)
        cv2.line(frame,(bx1,by),(bx2,by),(255,0,0),2)
        while int(ay) <= int((y+y+h)/2):
            if int(by) <= int((y+y+h)/2)&int(by+10) >= int((y+y+h)/2):
                cv2.line(frame,(bx1,by),(bx2,by),(0,255,0),2)
                Speed = Speed_Cal(time.time() - start_time)
                print("Car Number "+str(i)+" Speed: "+str(Speed))
                i = i + 1
                cv2.putText(frame,'Car', (x,y), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,0 ),1)
                cv2.putText(frame, "Speed: "+str(Speed)+"KM/H", (x,y-15), cv2.FONT_HERSHEY_PLAIN, 1, (255,0,0),1)

                
                break
            else :
                
                cv2.putText(frame, "Calcuting", (100,200), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0),3)
                break
                
       
        cv2.putText(frame,'Car', (x,y), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,0 ),1)
        blur = cv2.GaussianBlur(gray, (5, 5), 0)
        canny = cv2.Canny(blur, 10, 70)
        ret, mask = cv2.threshold(canny, 70, 255, cv2.THRESH_BINARY)
        cv2.imshow('Video feed', mask)

       
        fgmask1 = fgbg1.apply(frame)
        cv2.imshow('MOG', fgmask1)
        

    cv2.imshow('frame', frame)
   
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
